```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy.linalg as la
# Import the data set:
iris = pd.read_csv('iris.data', names=['sep_length', 'sep_width', 'pet_length', 'pet_width', 'class'])
print('iris dataset shape: ', iris.shape)
original = iris.to_numpy()
temp = iris.to_numpy()
```

```python
for i in range(4):
    original[:, i] = original[:, i] - np.mean(original[:, i])
for i in range(4):
    temp[:, i] = temp[:, i] - np.mean(temp[:, i])
```

```python
for i in range(100):
    scale = np.random.standard_cauchy()
    random_point = np.random.randint(0, iris.shape[0])
    random_index = np.random.randint(0, 4)
    temp[random_point][random_index] += scale
```

```python
plt.plot(temp[:, 1], temp[:, 3], "ro")
```

```python
def support(X, val, q): # Set q to be the qth quantile
    return np.quantile((X.T).dot(val), q)

def compute_fb(X, q, n):
    polar_body = np.ones((n,0))
    for i in range(X.shape[1]):
        if support(X, X[:,i] / la.norm(X[:,i]), q) > np.dot(X[:,i], X[:,i]/la.norm(X[:,i])): 
            polar_body = np.hstack((polar_body, np.array(X[:,i]).reshape(n,1)))
    return polar_body
```

```python
Samples = np.array([temp[:,1], temp[:,3]])
Samples_other = np.array([temp[:,0], temp[:,2]])
```

```python
fb = compute_fb(Samples, .95, 2)
fb_1 = compute_fb(Samples_other, .95, 2)
```

```python
plt.plot(original[:, 1], original[:, 3], "bo")
plt.show()
plt.plot(fb[0,:], fb[1,:], "ro")
plt.show()
```

```python
plt.plot(original[:, 0], original[:, 2], "bo")
plt.show()
plt.plot(fb_1[0,:], fb_1[1,:], "ro")
plt.show()
```
